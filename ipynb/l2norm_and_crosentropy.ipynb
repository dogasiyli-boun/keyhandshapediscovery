```python
import numpy as np
import torch
```

```python
#bt = np.asarray([[.05,.9,.05],[.6,.2,.2]])
bt = np.random.rand(10,4)
n,d=np.shape(bt)
print(n,' samples, ', d, ' dimensions')
print("bt:", bt)
print("bt square", ((bt * bt)))
print("bt square other", ((bt * bt))**2)
```

```python
l2_err = np.sqrt(np.sum(((bt * bt)) ** 2, 0))
print("activation of nodes\n", bt)
print("calc l2 error", l2_err)
print("mean l2 error", np.mean(l2_err))
print("sum l2 error", np.sum(l2_err))
```

```python
sq_err_per_dim_sample = bt*bt
print("sq_err_per_dim_sample:")
print(sq_err_per_dim_sample)
```

```python
print(np.sqrt(np.sum(sq_err_per_dim_sample, 0)))
```

```python
loss_CLS_fun = torch.nn.CrossEntropyLoss(reduction='mean')
bt_t = torch.Tensor(bt)
_, preds = torch.max(bt_t, 1)
print("bt:", bt_t)
print("preds:", preds)
CLS = loss_CLS_fun(bt_t, preds)
print(CLS.item())
```
